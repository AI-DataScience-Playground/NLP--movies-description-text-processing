```python
import string
import gensim
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

nltk.download('stopwords')
```

```python
def read_file_content(filename):
    with open(filename, 'r') as file:
        return file.read()

kfp_text = read_file_content('KFP1Script.csv') + read_file_content('KFP2Script.csv')
kfp_sentences = sent_tokenize(kfp_text)
print(kfp_sentences)
```

```python
final_dataset = []
for sent in kfp_sentences:
    stopwords_list = set(stopwords.words('english'))
    sent = sent.translate(str.maketrans('','',string.punctuation))
    kfp_preprocessed_list = []
    for word in sent.split():
        word = word.lower()
        if(word not in stopwords_list):
            kfp_preprocessed_list.append(word)
    final_dataset.append(kfp_preprocessed_list)

print(final_dataset)
```

```python
print(len(final_dataset))
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)

model.build_vocab(final_dataset)

print('complete!')
```

```python
model.train(final_dataset, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
vocab = list(model.wv.index_to_key)
print("Vocabulary:", vocab)
```
